```python
import requests
import os

# Define the URL and the target directory
url = "https://huggingface.co/ai-forever/Real-ESRGAN/resolve/main/RealESRGAN_x4.pth?download=true"
target_dir = "/kaggle/working/ComfyUI/models/upscale_models"

# Create the target directory if it doesn't exist
os.makedirs(target_dir, exist_ok=True)

# Define the file name
filename = os.path.join(target_dir, "RealESRGAN_x4.pth")

# Download the file
response = requests.get(url)
if response.status_code == 200:
    # Write the downloaded content to the file
    with open(filename, 'wb') as f:
        f.write(response.content)
    print(f"File downloaded successfully: {filename}")
else:
    print(f"Failed to download file from {url}. Status code: {response.status_code}")
```
